```python
#OpenCv
import cv2
from google.colab.patches import cv2_imshow
```

```python
X= cv2.imread('flor.jpeg') #BRG
cv2_imshow(X)
```

```python
#Canal azul
B=X[:,:,0]
cv2_imshow(B)
```

```python
#Canal verde
G=X[:,:,1]
cv2_imshow(G)
```

```python
#Canal rojo
R=X[:,:,2]
cv2_imshow(R)
```

##Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
X=plt.imread('flor.jpeg') #RGB
plt.imshow(X)
```

```python
#Canal rojo
R = X[:,:,0]
plt.ishow(R, cmap='gray')
```

##Opencv y Matplotlib

```python
X = cv2.imread('flor.jpeg')
X = cv2.cvtColor(X,cv2.COLOR_BGR2RGB)

plt.imshow(X)
```
