```python
#importación de lilbrerías
import numpy as np
from cv2 import imread
import matplotlib.pyplot as plt
```

### Aquí inicia procesamiento de la imagen:

```python
#lectura y desplieghue de la imagen
img=imread("onerice.jpg")
plt.imshow(img)
```

```python
#Entender imagen
def howis(img):
    print('size:', img.shape)
    print('max:', np.max(img))
    print('min:', np.min(img))
```

```python
howis (img)
```

```python
#Visualización x canales (primer canal)

X= img[:,:,0]
howis(X)
plt.imshow(X, cmap='gray')
plt.show()
```

```python
#se ve un poquito más oscuro
```

```python
#Fn para segmentación según un umbral

def segmentar(X,t):
    (N,M)=X.shape
    Y=np.zeros(X.shape)
    for i in range(N):
        for j in range(M):
            if X[i,j]>t:
                Y[i,j]=255
    return Y
```

```python
Y=segmentar(X,100)
howis(Y)
plt.imshow(Y, cmap='gray')
plt.show()
```

```python
#como se ven puntitos blancos:
Y=segmentar(X,130)
howis(Y)
plt.imshow(Y, cmap='gray')
plt.show()
```

```python
#Fn para segmentación según un umbral

def segmentar(X,t):
    (N,M)=X.shape #imagen 2D
    Y=np.zeros(X.shape) #crea imagen de salida del mismo tamaño que original (matriz 1 y 0s)
    area = 0 #inicializamos contador
    for i in range(N): #Recorren filas
        for j in range(M): #recorren columnas
            if X[i,j]>t: #comparación del pixel es > al umbral lo iguala a ;la escala 255
                Y[i,j]=255
                area = area+1
    print("area:", area)
    return Y
```

```python
Y=segmentar(X,130)
howis(Y)
plt.imshow(Y, cmap='gray')
plt.show()
```
